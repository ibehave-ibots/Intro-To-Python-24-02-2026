```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('gapminder.csv')
df.head()
```

```python
df = df[df['country'] == 'Germany']
plt.plot(df['year'], df['pop'], color='k', marker='.', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Population')
```

```python
df_ger = df[df['country'] == 'Germany']
df_usa = df[df['country'] == 'United States']

plt.plot(df_ger['year'], df_ger['pop']/1e6, color='k', marker='.', linestyle='--', label='Germany')
plt.plot(df_usa['year'], df_usa['pop']/1e6, color='b', marker='.', linestyle='-.', label='USA')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Population [Million]')
plt.title('Population over the years')
```
